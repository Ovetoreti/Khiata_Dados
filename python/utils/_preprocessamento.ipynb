### <h3> Preparação

- O formulário feito no Google trouxe uma coluna desnecessária, portanto, optamos por apaga-la

- Renomeamos os nomes das colunas do nosso dataset, pois todas possuiam nomes grandes.

- Verificamos o estado do DataFrame, e como não identificamos nada de incomum, seguimos para a etapa de Save

- Por fim, salvamos o novo DataFrame

<hr>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'FormsKhiataOriginal.csv')
```

##### Excluindo as colunas desnecessárias e as renomeando

```python
# Criando uma lista das colunas que irão ser excluidas do df
colunas_excluir = ['Carimbo de data/hora']

df = df.drop(colunas_excluir, axis=1)

df.rename(columns={
    "Digite a sua faixa etária": "Faixa_Etaria", 
    "Você já consumiu algum produto artesanal?": "Consumiu_Prod_Artesanal",
    "Qual seu nível de interesse em relação a produtos feitos à mão?": "Nvl_Interesse_Prod_Artesanal",
    "Com que frequência você compra produtos artesanais?": "Freq_Compra_Prod_Artesanal",
    "Você já utilizou um aplicativo/plataforma para comprar produtos artesanais?": "Utilizou_App_Prod_Artesanal",
    "Qual tipo de produto artesanal você mais se interessa?": "Tipo_Produto_Mais_Interessado",
    "Qual a faixa de preço que você estaria disposto(a) a pagar por um produto artesanal?": "Faixa_Preco_Interessado",
    "Utilizaria um app que une o cliente a costureira?": "Utilizaria_App_Costura"}, inplace=True)
```

##### Análise da base

```python
# Informações básicas do dataset
print(f'Dataset contém {df.shape[0]} linhas e {df.shape[1]} colunas')

# Verificar valores ausentes
print('\nValores ausentes por coluna:')
print(df.isnull().sum())

# Estatísticas descritivas para variáveis numéricas
print('\nEstatísticas descritivas para variáveis numéricas:')
display(df.describe())

# Estatísticas descritivas para variáveis categóricas
print('\nEstatísticas descritivas para variáveis categóricas:')
display(df.describe(include='O'))

# Verificar correlação entre variáveis numéricas
print('\nMatriz de correlação entre variáveis numéricas:')

# Selecionar apenas colunas numéricas
numeric_columns = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr()

display(correlation_matrix)
```

##### Salvando a nova base

```python
# Salvando o Dataframe preprocessado em .CSV
df.to_csv('FormsKhiataTratada.csv', index=False)  # index=False evita salvar o índice como uma coluna extra
```
