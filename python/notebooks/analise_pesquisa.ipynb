Análise de público-alvo feita com base no formulário enviado com 8 perguntas relacionadas a costura

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\enzorocha-ieg\OneDrive - Instituto Germinare\Área de Trabalho\Khiata_Dados\python\utils\FormsKhiataTratada.csv')

# Ajustar estilo
sns.set(style="whitegrid")
```

Gráfico 1: Distribuição de Faixa Etária

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Faixa_Etaria", order=df["Faixa_Etaria"].value_counts().index, palette="viridis")
plt.title("Distribuição de Faixa Etária")
plt.xlabel("Faixa Etária")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```

Gráfico 2: Frequência de Compra de Produtos Artesanais

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Freq_Compra_Prod_Artesanal", order=df["Freq_Compra_Prod_Artesanal"].value_counts().index, palette="viridis")
plt.title("Frequência de Compra de Produtos Artesanais")
plt.xlabel("Frequência de Compra")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```

Gráfico 3: Tipo de Produto Mais Interessado

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Tipo_Produto_Mais_Interessado", order=df["Tipo_Produto_Mais_Interessado"].value_counts().index, palette="viridis")
plt.title("Tipo de Produto de Maior Interesse")
plt.xlabel("Tipo de Produto")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```

Gráfico 4: Faixa de Preço Interessado

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Faixa_Preco_Interessado", order=df["Faixa_Preco_Interessado"].value_counts().index, palette="viridis")
plt.title("Faixa de Preço de Interesse")
plt.xlabel("Faixa de Preço")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```
